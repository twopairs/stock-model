```python
# yahoo 에서 data 확보
# modeling을 종가로 try
# import yfinance as yf

# stock_id = "005930.KS"
# stock_id = "035420.KS"
# stock_id = "035720.KS"
stock_id = "017670.KS"

import pandas as pd
import yfinance as yf

pd.set_option('display.float_format', '{:.2f}'.format)  # 소수 둘째자리까지


data = yf.Ticker(stock_id)
hist_data = data.history()
hist_data_period = data.history(start="2025-01-01").reset_index()
hist_data_period
```

```python
from datetime import datetime, timedelta
start = datetime(2025, 10, 1)
hist_data_period['Date'] = hist_data_period['Date'].dt.tz_localize(None)

df = hist_data_period[hist_data_period['Date'] > start]
df
```

```python
def analysis_profit(_df):
    _count = len(_df)
    _invest = _df['Close'].sum()
    _last_row = _df.iloc[-1]
    _price = _last_row['Close']
    _profit = _count * _price - _invest
    _profit_rate = _profit * 100 / _invest
    return _profit, _profit_rate
```

```python

now = datetime.now()
s = start
delta = (now - s).days

result = []

for x in range(delta):
    e = s + timedelta(days=x)

    df_sub = df[(df['Date'] > s) & (df['Date'] <= e)]

    if len(df_sub) < 3:
        continue

    profit, profit_rate = analysis_profit(df_sub)

    # print(s, e, profit, profit_rate)

    result.append({
        "s": s,
        "e": e,
        "profit": profit,
        "profit_rate": profit_rate
    })


df_ret = pd.DataFrame(result)
df_ret

```

```python

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(40,10))
ax = sns.barplot(
    data=df_ret,
    x="e",
    y="profit"        # 기본값이 count라 생략해도 됨
)

plt.xticks(rotation=45)
plt.show()


```
